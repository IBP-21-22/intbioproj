```python
data = pd.read_excel('../raw_data/41592_2018_45_MOESM4_ESM.xlsx', skiprows=6)
data.head()
```

```python
genes = data['Standard Name'].dropna()
```

```python
genes.duplicated().any()
```

```python
genes
```

```python
ORG = 'sce' # yeast
```

```python
import requests
```

```python
from tqdm.notebook import tqdm
```

```python
ident1 = 'PATHWAY'
ident2 = 'BRITE'
raw_texts = {}
pathways = {}
for gene in tqdm(genes):
    print(gene)
    query = f'http://rest.kegg.jp/get/{ORG}:{gene}/'
    ret = requests.get(query).text
    raw_texts[gene] = ret
    try:    
        pathway_content = ret[ret.find(ident1) + len(ident1):ret.find(ident2)].rstrip().strip()
        pathways[gene] = [x.strip() for x in pathway_content.split('\n')]
    except:
        print("Error")
        pass
        
```

```python
import pickle
with open('../processed_data/raw_pathways.pkl','wb') as out:
    pickle.dump(raw_texts, out)
```

```python
pathwaysDf = pd.Series(pathways).reset_index()
pathwaysDf.columns = ['Gene', 'pathways']
```

```python
pathwaysDf.to_csv('../processed_data/genesMappingPW.csv', index=False)
```
