```python
raw_texts = {}
try:
    with open('../raw_data/raw_pathways.pkl','rb') as inp:
        raw_texts = pickle.load(inp)
except IOError:
    for gene in tqdm(genes):
        query = f'http://rest.kegg.jp/get/{ORG}:{gene}/'
        ret = requests.get(query).text
        raw_texts[gene] = ret
    import pickle
    with open('../raw_data/raw_pathways.pkl','wb') as out:
        pickle.dump(raw_texts, out)
```

```python
ident1 = 'PATHWAY'
pathways = {}
for gene, ret in raw_texts.items():
    lines = ret.split('\n')
    try:
        start_line_ind = next(cnt for cnt,x in enumerate(lines) if ident1 in x)
    except StopIteration:
        print(f"{gene}: No Pathways Information")
        assert ident1 not in ret
        continue
    try:
        end_line_ind = start_line_ind + next(
            cnt for cnt,x in enumerate(lines[start_line_ind + 1:]) if not x.startswith(' '))
    except StopIteration:
        print(lines)
        print(start_line_ind)
        print(lines[start_line_ind:])
        raise
    pathway_content = lines[start_line_ind:end_line_ind + 1]
    pathways[gene] = [x.strip() for x in pathway_content]
    
```

```python
pathwaysDf = pd.Series(pathways).reset_index()
pathwaysDf.columns = ['Gene', 'pathways']
```
